```python
import pandas as pd
import seaborn as sns

# Create a dummy shipments dataset
valor_do_frete = [200, 300, 150, 350, 250, 200, 100, 300, 250, 300, 200, 100, 150, 200, 250]
competitor = ["A", "B", "C"] * 5
freight_type = ["F1", "F2", "F3", "F4", "F5"] * 3


shipments = pd.DataFrame({
"valor_do_frete": valor_do_frete,
"competitor": competitor,
"freight_type": freight_type
})

# Set the seaborn style to whitegrid
sns.set_theme(style="whitegrid")

# Draw a pointplot to show shipment volume as a function of two categorical factors
g = sns.catplot(
data=shipments, x="freight_type", y="valor_do_frete", hue="competitor",
capsize=.2, palette="tab10",
kind="point", height=6, aspect=2, errorbar="ci"
)
g.despine(left=True)
```
