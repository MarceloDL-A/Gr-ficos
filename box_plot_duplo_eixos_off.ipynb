```python
import seaborn as sns
import matplotlib.pyplot as plt

# exemplo de dataframe
df = sns.load_dataset("titanic")

fig, ax = plt.subplots()
sns.boxplot(data=df, x="age", y="class", hue="alive", ax=ax)

# Adicionar anotações com número de instâncias
for i in range(len(df.groupby("class"))):
    for j, alive in enumerate(df["alive"].unique()):
        n = len(df[(df["class"] == df["class"].unique()[i]) & (df["alive"] == alive)])
        ax.annotate(f"{n}", xy=(df.age.max() + 4, i - 0.25 + (j * 0.25)), xycoords='data', ha='left', va='center')

# Desativar bordas e grade
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)


plt.show()
```
