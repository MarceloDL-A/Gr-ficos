```python
import re
import unicodedata
import pandas as pd
import numpy as np

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    only_ascii = nfkd_form.encode('ASCII', 'ignore')
    return only_ascii.decode()
def remove_non_letters(string):
    return re.sub(r'[^a-zA-Z]', '', string)

def regex_treatment(df, col_name):
    df[col_name] = df[col_name].apply(remove_accents)
    df[col_name] = df[col_name].str.upper()
    df[col_name] = df[col_name].apply(remove_non_letters)
    
    return df

product_names = ['Produto A', 'd;khfsa; B', 'Prodkjaslkdfgjaáééíuto C', 'Produto D', 'Produto E', 'Produto F', 'Produto G', 'Produto H', 'Produto I', 'Produto J']
quantities = [2, 3, 5, 6, 7, 4, 6, 7, 5, 4]
df = pd.DataFrame({'Product Name': product_names, 'Quantity': quantities})
df = regex_treatment(df, 'Product Name')

df
```
