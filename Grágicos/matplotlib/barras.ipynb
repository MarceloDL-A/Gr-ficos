```python
import matplotlib.pyplot as plt
import pandas as pd

# Carregue seu dataframe
df = pd.read_csv('frete_promessa_AMERICANAS.csv')
```

```python
df.head()
```

```python
# Crie as barras de erro
erro = df['FREIGHT_PRICE_STD']

# Crie o gráfico de barras
plt.bar(df['GRUPO'], df['FREIGHT_PRICE_AVG'], yerr=erro, capsize=5)
plt.bar(df['GRUPO'], df['FREIGHT_PRICE_AVG'], yerr=erro, capsize=5)

# Adicione rótulos e títulos
plt.xlabel('Grupos')
plt.ylabel('Médias')
plt.title("""Concorrente: Americanas
Produto: Detergente Líquido Clear Ypê 500ml
Promessa: 6 dias
São desconsiderados os FREE SHIPPINGS""", loc='left')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
ax.bar(df["GRUPO"], df["FREIGHT_PRICE_AVG"], yerr=df["FREIGHT_PRICE_STD"], label="concorrente 1")
ax.bar(df["GRUPO"], df["FREIGHT_PRICE_AVG"], yerr=df["FREIGHT_PRICE_STD"], label="concorrente 2")
ax.set_xlabel("Grupos")
ax.set_ylabel("Média")
ax.set_title("Histograma com barra de erro")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
bar_width = 0.4
left_pos = np.arange(len(df["GRUPO"]))
ax.barh(left_pos, df["FREIGHT_PRICE_AVG"], bar_width,  label="concorrente 1", xerr = df["FREIGHT_PRICE_STD"])
ax.barh(left_pos + bar_width, df["FREIGHT_PRICE_AVG"], bar_width, label="concorrente 2", xerr = df["FREIGHT_PRICE_STD"])
ax.set_yticks(left_pos + bar_width / 2)
ax.set_yticklabels(df["GRUPO"])
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
bar_width = 0.4
left_pos = np.arange(len(df["GRUPO"]))
ax.bar(left_pos, df["FREIGHT_PRICE_AVG"], bar_width,  label="concorrente 1", yerr = df["FREIGHT_PRICE_STD"], capsize = 2)
ax.bar(left_pos + bar_width, df["FREIGHT_PRICE_AVG"], bar_width, label="concorrente 2", yerr = df["FREIGHT_PRICE_STD"], capsize = 2)
ax.set_xticks(left_pos + bar_width / 2)
ax.set_xticklabels(df["GRUPO"])
ax.legend(loc = 'upper left')
plt.show()
```
