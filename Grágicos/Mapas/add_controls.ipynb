```python
df_LATLONG = df_coordenadas[['NM_MUNICIPIO', 'LONG', 'LAT']]
```

```python
df_LATLONG
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
df_LATLONG.columns
```

```python
df = df_LATLONG.sample(n=100)
```

# Exemplo 1 de Folium

```python
mapc = folium.Map(location=[df_LATLONG.LAT.mean(), df_LATLONG.LONG.mean()], zoom_start=5, control_scale=True)

# Configuração de heatmap
heatmap = HeatMap(list(zip(df.LAT, df.LONG, df["LONG"]**10)),
               min_opacity=0.5,
               max_val=df["LONG"].max(),
               radius=50, blur=30, 
               max_zoom=1,
              control=True,
        )
# Cria camadas
layer1 = folium.FeatureGroup(name='Markers', control=True)
layer2 = folium.FeatureGroup(name='Circulos', control=True)
layer3 = folium.FeatureGroup(name='Heatmap', control=True)
for index, location_info in df.iterrows():

    # Adiciona propriedades às camadas
    layer1.add_child(folium.Marker([location_info["LAT"], location_info["LONG"]], popup=location_info["NM_MUNICIPIO"]))
    layer2.add_child(folium.CircleMarker([location_info["LAT"], location_info["LONG"]], radius=(abs(location_info["LAT"])), fill_color='blue', popup=location_info["NM_MUNICIPIO"]))


layer3.add_child(heatmap)

# Adiciona as camadas ao mapa
mapc.add_child(layer1)
mapc.add_child(layer2)
mapc.add_child(layer3)


# Adiciona o controle de camdas ao mapa
folium.LayerControl().add_to(mapc)
```
