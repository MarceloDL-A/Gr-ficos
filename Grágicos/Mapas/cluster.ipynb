# Exemplo 3 de Folium

```python
from folium import plugins
```

```python
m = folium.Map(location=[df.LAT.mean(), df.LONG.mean()], zoom_start=5,width="%100",height="%100",)
locations = list(zip(df["LAT"], df["LONG"]))
cluster = plugins.MarkerCluster(locations=locations,                     
               popups=df["NM_MUNICIPIO"].tolist(), popup =  df["NM_MUNICIPIO"].tolist())  
m.add_child(cluster)
m
```
