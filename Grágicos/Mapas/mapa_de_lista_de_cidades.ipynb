```python
from DadosAbertosBrasil import ibge
import folium
import pandas as pd

def plot_cities_on_map(cities):
    df_localidades = pd.DataFrame(ibge.coordenadas())
    df_localidades = df_localidades[df_localidades['NM_DISTRITO'].isin([i.upper() for i in cities])]
    m = folium.Map(location=[-15.77972, -47.92972], zoom_start=4)
    for index, row in df_localidades.iterrows():
#         folium.Marker(location=[row['LAT'], row['LONG']],popup=row['NM_DISTRITO']).add_to(m)
        folium.CircleMarker(location=[row['LAT'], row['LONG']],radius=3,
                            color='#3186cc',fill_color='#3186cc',popup=row['NM_DISTRITO']).add_to(m)
    return m

# Example usage
cities = ["SÃ£o Paulo", "Rio de Janeiro", "Belo Horizonte", "Salvador"]
mapa = plot_cities_on_map(cities)
mapa
```
