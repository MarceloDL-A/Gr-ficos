```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Define os 3 números
n1 = set(["A", "B", "C"])
n2 = set(["A", 7, 15])
n3 = set([10])

# Crie o diagrama de Venn
venn = venn3([n1, n2, n3], set_labels = ('Conjunto 1', 'Conjunto 2', 'Conjunto 3'))

# Mostre o gráfico
plt.show()
```
